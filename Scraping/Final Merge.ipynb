```python
import pandas as pd
```

```python
df = pd.read_csv("res3.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(["Unnamed: 6"],axis = 1)
```

```python
for i in df.iterrows():
    df.at[i[0], "Brand"] = str(i[1]["Brand"]).title()
    df.at[i[0], "Category"] = str(i[1]["Category"]).title()
    if i[1]["Country"] == "Indian":
        df.at[i[0], "Country"]  = "India"
    
```

```python
df[df["Country"]=="Indian"]
```

```python
df2 = pd.read_csv("brand.csv",encoding='latin1')
```

```python
df2.head()
```

```python
df2.shape
```

```python
for i in df2.iterrows():
    df2.at[i[0], "Brand"] = str(i[1]["Brand"]).title()
    df2.at[i[0], "Category"] = str(i[1]["Category"]).title()
    
```

```python
result = pd.concat([df,df2])
```

```python
result
```

```python
dup = set()
for i in result.iterrows():
    if len(result[result["Brand"] == i[1]["Brand"]])>1:
        dup.add(i[1]["Brand"])
```

```python
dup
```

```python
result = result.reset_index(drop = True)
```

```python
res = result.drop_duplicates(subset='Brand', keep="last").reset_index(drop=True)
```

```python
res
```

```python
res.to_csv("output.csv",index=False)
```
